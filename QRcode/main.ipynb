```python
from tkinter import *
import qrcode
from PIL import Image,ImageTk
from resizeimage import resizeimage
```

```python
class qrgen:
    def __init__(self,root):
        self.root=root
        self.root.geometry("640x500+300+100")
        self.root.title("QR_generator_by_phoenix")
        self.root.resizable(False,False)
        
        title=Label(self.root,text="QR Code Generator",font=("times new roman",40),bg='skyblue',fg='white').place(x=0,y=0,relwidth=1)
        
        #Employee detailed window
        #variables
        self.var_emp_code=StringVar()
        self.var_emp_name=StringVar()
        self.var_emp_dept=StringVar()
        self.var_emp_designation=StringVar()
        
        emp_Frame=Frame(self.root,bd=2,relief=RIDGE,bg='white')
        emp_Frame.place(x=50,y=100,width=300,height=350)
        emp_title=Label(emp_Frame,text="Employee Details",font=("goudy old style",20),bg='black',fg='white').place(x=0,y=0,relwidth=1)
        
        lbl_emp_code=Label(emp_Frame,text="Employee ID :",font=("timesnew roman",10,'bold'),fg='black').place(x=20,y=50)
        lbl_emp_name=Label(emp_Frame,text="Employee Name :",font=("timesnew roman",10,'bold'),fg='black').place(x=20,y=100)
        lbl_emp_dept=Label(emp_Frame,text="Department :",font=("timesnew roman",10,'bold'),fg='black').place(x=20,y=150)
        lbl_emp_designation=Label(emp_Frame,text="Designation :",font=("timesnew roman",10,'bold'),fg='black').place(x=20,y=200)
        
        txt_emp_code=Entry(emp_Frame,font=("timesnew roman",10),textvariable=self.var_emp_code,bg='lightyellow').place(x=140,y=50)
        txt_emp_name=Entry(emp_Frame,font=("timesnew roman",10),textvariable=self.var_emp_name,bg='lightyellow').place(x=140,y=100)
        txt_emp_dept=Entry(emp_Frame,font=("timesnew roman",10),textvariable=self.var_emp_dept,bg='lightyellow').place(x=140,y=150)
        txt_emp_designation=Entry(emp_Frame,font=("timesnew roman",10),textvariable=self.var_emp_designation,bg='lightyellow').place(x=140,y=200)
        
        btn_generate=Button(emp_Frame,text="QR Generate",command=self.generate,font=("times new roman",15),bg='skyblue',fg='white').place(x=20,y=250)
        btn_clear=Button(emp_Frame,text="Clear",command=self.clear,font=("times new roman",15),bg='gray',fg='white').place(x=180,y=250)
        
        self.msg=''
        self.lbl_msg=Label(emp_Frame,text=self.msg,font=("timesnew roman",15),bg='white',fg='green')
        self.lbl_msg.place(x=40,y=300)
        
        #Employee qrcode window
        qr_Frame=Frame(self.root,bd=2,relief=RIDGE,bg='white')
        qr_Frame.place(x=360,y=100,width=240,height=240)
        qr_title=Label(qr_Frame,text="QR code",font=("goudy old style",20),bg='black',fg='white').place(x=0,y=0,relwidth=1)
        
        self.qr_code=Label(qr_Frame,text='No QR\nAvilable',font=('times new roman',10),bg='blue',fg='white',bd=1,relief=RIDGE)
        self.qr_code.place(x=80,y=80,width=100,height=100)
    
    def clear(self):
        self.var_emp_code.set('')
        self.var_emp_name.set('')
        self.var_emp_dept.set('')
        self.var_emp_designation.set('')
        self.msg=''
        self.lbl_msg.config(text=self.msg)
        self.qr_code.config(image='')
        
    def generate(self):
        if self.var_emp_designation.get()==''or self.var_emp_code.get()==''or self.var_emp_name.get()==''or self.var_emp_dept.get()=='':
            self.msg='All Field are Required'
            self.lbl_msg.config(text=self.msg,fg='red')
        else:
            qr_data=(f"Employee ID :{self.var_emp_code.get()}\nEmployee Name :{self.var_emp_name.get()}\nEmployee Department :{self.var_emp_dept.get()}\nEmployee designation :{self.var_emp_designation.get()}")
            qr_code=qrcode.make(qr_data)
            #print(qr_code)
            qr_code=resizeimage.resize_cover(qr_code,[100,100])
            qr_code.save('QR_codes/Emp_'+str(self.var_emp_code.get())+'.png')
            #qrcode image update
            self.im=ImageTk.PhotoImage(file='QR_codes/Emp_'+str(self.var_emp_code.get())+'.png')
            self.qr_code.config(image=self.im)
            #updating notification
            self.msg='QR generated succesfully!!'
            self.lbl_msg.config(text=self.msg,fg='green')
        
root=Tk()
obj=qrgen(root)
root.mainloop()
```
